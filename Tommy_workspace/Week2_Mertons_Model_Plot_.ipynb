```
import numpy as np
import matplotlib.pyplot as plt
```

```
V_0     = 100
mu_V    = 0.05
sigma_V = 0.2
T       = 1.0
B       = 90
N       = 1000

dt = T / N

np.random.seed(42)
```

```
dW = np.random.normal(loc=0, scale=np.sqrt(dt), size=N)
W  = np.concatenate([[0], dW.cumsum()])


time = np.linspace(0, T, N + 1)

V = V_0 * np.exp((mu_V - 0.5 * sigma_V**2) * time + sigma_V * W)
```

```
plt.figure(figsize=(7, 4))

plt.plot(time, V, color="steelblue", linewidth=0.8, label="Path of $V_T$")
plt.axhline(y=B, color="red", linestyle="--", linewidth=1.2,
            label=f"Default Threshold B={B}")

plt.title("Simulated Path of $V_T$")
plt.xlabel("Time")
plt.ylabel("$V_t$")
plt.legend()
plt.tight_layout()


plt.show()
```
