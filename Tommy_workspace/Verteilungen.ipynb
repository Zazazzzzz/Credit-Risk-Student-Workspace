```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# ----------------------------
# Parameter (Credit Risk Setup)
# ----------------------------
n = 50          # Anzahl Kredite im Portfolio
p = 0.1          # Ausfallwahrscheinlichkeit (PD)
lambda_ = n * p   # Erwartete Anzahl Defaults (Poisson Parameter)

# Wertebereich für Anzahl Defaults
x = np.arange(0, 40)

# ----------------------------
# Wahrscheinlichkeiten berechnen
# ----------------------------
binom_pmf = binom.pmf(x, n, p)
poisson_pmf = poisson.pmf(x, lambda_)

# ----------------------------
# Plot
# ----------------------------
plt.figure()

plt.plot(x, binom_pmf, marker='o', linestyle='-', label='Binomialverteilung')
plt.plot(x, poisson_pmf, marker='x', linestyle='--', label='Poisson-Approximation')

plt.xlabel('Anzahl Ausfälle (Defaults)')
plt.ylabel('Wahrscheinlichkeit')
plt.title('Vergleich: Binomial vs. Poisson (Credit Risk)')
plt.legend()
plt.grid()

plt.show()
```
